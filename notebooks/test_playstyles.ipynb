```python
import sys
sys.path.append("..")  

import pandas as pd
from ml.playstyles import compute_playstyle_clusters
from ml.features import college_features  # or whichever feature list you want
```

```python
# Load your latest draftpool
df = pd.read_csv("../data/processed/2024/draftpool_stats_2024.csv")

# Filter to College prospects for consistent features
college_df = df[df['classification'] == 'College']

# Run clustering (can adjust n_clusters, etc.)
clustered_df = compute_playstyle_clusters(
    college_df,
    feature_cols=college_features,
    n_clusters=5,
    use_pca=True,
    pca_components=2,
    plot=True
)

clustered_df
```
