```python
# 2021
teams_df_2021 =  pd.read_csv("../data/raw/team_stats_2021.csv")
teams_df_2021["Team"] = teams_df_2021["Team"].str.replace('*', '')
teams_df_2021
```

```python
# 2020
teams_df_2020 =  pd.read_csv("../data/raw/team_stats_2020.csv")
teams_df_2020["Team"] = teams_df_2020["Team"].str.replace('*', '')
teams_df_2020
```

```python
team_dfs = {
    2020: teams_df_2020,
    2021: teams_df_2021,
    2022: teams_df_2022,
    2023: teams_df_2023,
    2024: teams_df_2024
}
```

```python
def merge_draft_with_team(year):
    try:
        draft_df = pd.read_csv(f"../data/raw/draft_{year}.csv")
        team_df = team_dfs[year].copy()
        
        # drop unnecessary cols, add win %
        team_df.drop(columns=["Unnamed: 22", "Unnamed: 27", "Arena", "Attend.", "Attend./G"], inplace=True, errors='ignore')
        team_df["win_pct"] = team_df["W"] / (team_df["W"] + team_df["L"])

        # merge draft order with team stats
        merged_df = pd.merge(draft_df, team_df, left_on="team", right_on="Team", how="left")
        merged_df.drop(columns=["Team"], inplace=True, errors='ignore')

        merged_df.to_csv(f"../data/processed/draft_{year}_team_stats.csv", index=False)
    except Exception as e:
        print(f"Error processing year {year}: {e}")

for year in range(2020, 2025):
    merge_draft_with_team(year)
```
